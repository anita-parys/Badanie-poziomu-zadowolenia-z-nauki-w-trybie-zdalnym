##Modele regresji linowej 

```python
tab3=tab.groupby(['Ogólny poziom zadowolenia ze studiów w trybie online'], as_index = True).mean().round(2).sort_values('Ogólny poziom zadowolenia ze studiów w trybie online')
s=tab3[['Czy możesz utrzymać skupienie podczas zajęć online']]
s=s.to_numpy()

N = 15
x =  list(range(1, 11))
y = s
# fit a linear curve an estimate its y-values and their error.
a, b = np.polyfit(x, y, deg=1)
y_est = a * x + b
y_err = np.std(x) * np.sqrt(1/len(x) +
                          (x - np.mean(x))**2 / np.sum((x - np.mean(x))**2))

fig, ax = plt.subplots(figsize = (16,10))
ax.plot(x, y_est, '-')
ax.set_xlabel('Ogólny poziom zadowolenia z nauki w trybie online')
ax.set_ylabel('Skupienie podczas nauki w trybie online')

ax.set_title('Wykres zależności między ogólnym poziomem zadowolenia a utrzymywanym skupieniem')
ax.fill_between(x, y_est - y_err, y_est + y_err, alpha=0.2)
ax.plot(x, y, 'o', color='tab:red')
```

```python
s=tab3[['Czy Twoim zdaniem nauka w tym modelu nauczania jest skuteczniejsza w porównaniu do modelu klasycznego']]
s=s.to_numpy()
N = 15
x =  list(range(1, 11))
y = s
# fit a linear curve an estimate its y-values and their error.
a, b = np.polyfit(x, y, deg=1)
y_est = a * x + b
y_err = np.std(x) * np.sqrt(1/len(x) +
                          (x - np.mean(x))**2 / np.sum((x - np.mean(x))**2))

fig, ax = plt.subplots(figsize = (16,10))
ax.plot(x, y_est, '-')

ax.set_xlabel('Ogólny poziom zadowolenia z nauki w trybie online')
ax.set_ylabel('Ocena modelu nauki w trybie zdalnym')

ax.set_title('Wykres zależności między ogólnym poziomem zadowolenia a oceną modelu nauki w trybie zdalnym')
ax.fill_between(x, y_est - y_err, y_est + y_err, alpha=0.1, color ='green')
ax.plot(x, y, 'o', color='green')
```
